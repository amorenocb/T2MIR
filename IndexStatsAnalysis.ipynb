```python
import pandas as pd
%matplotlib inline
import matplotlib.pyplot as plt
plt.rc('font', size=20)
plt.rc('legend', fontsize=20)
plt.rc('figure', titlesize=20)
plt.rc('axes', titlesize=20)
plt.rc('axes', labelsize=20) 
plt.rc('xtick', labelsize=15)    
plt.rc('ytick', labelsize=15)
```

```python
pca_stats = pd.read_csv("pca-stats.csv")
non_pca_stats = pd.read_csv("non-pca-stats.csv")
```

```python
pca_stats.head(5)
```

```python
non_pca_stats.tail(5)
```

```python
sub_plots = pca_stats.groupby(['Features'])
for group, data in sub_plots:
    plt.figure(figsize=(5,10))
    lines = data.groupby(['Algoritmo','Parametro'])
    for group2, line in lines:
        sorted_line = line.sort_values(by='Checks')
        plt.plot(sorted_line['Efectividad'], sorted_line['Eficiencia'], '-o' ,label="{}-{}".format(group2[0],group2[1]))
    
    plt.title(group)
    plt.xlabel('Efectividad')
    plt.ylabel('Eficiencia')
    plt.legend(loc='best')
    plt.show()
```

```python
sub_plots = non_pca_stats.groupby(['Features'])
for group, data in sub_plots:
    plt.figure(figsize=(5,10))
    lines = data.groupby(['Algoritmo','Parametro'])
    for group2, line in lines:
        sorted_line = line.sort_values(by='Checks')
        plt.plot(sorted_line['Efectividad'], sorted_line['Eficiencia'], '-o',label="{}-{}".format(group2[0],group2[1]))
    
    plt.title(group)
    plt.xlabel('Efectividad')
    plt.ylabel('Eficiencia')
    plt.legend(loc='best')
    plt.show()
```
